```python
df_rest_avg = df_restaurants.pivot_table(
                            values = ['popularity','rating','price'],
                            index = ['Postal_code'],
                            aggfunc=np.mean
                            )
df_rest_avg
```

```python
df_rest_count = df_restaurants.pivot_table(
                                index = ['Postal_code'],
                                values = ['category'],
                                aggfunc = 'count'
                                    )
df_rest_count
```

```python
df_ind_rest_count = df_restaurants[df_restaurants['category'].str.contains("Indian")].pivot_table(
                                                                            index = ['Postal_code'],
                                                                            values = ['category'],
                                                                            aggfunc = 'count'
                                                                            )
df_ind_rest_count
```
