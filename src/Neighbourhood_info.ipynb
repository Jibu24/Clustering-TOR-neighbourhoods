```python
# finding neighborhood names from wikipedia
html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")
df_wiki = pd.read_html(html)[0]
```

```python
df_wiki.head()
```

```python
# The neighborhood names are provided inside the brackets, we need to extract them and the corresponding postal code
from data_processing.functions import extract_neighborhood_name
df_hoods = extract_neighborhood_name(df_wiki)
df_hoods.head()
```

```python
# Merge both dataframes and drop any neighbourhoods we dont have lat and lon coordinates for
df_neighbourhoods = df.merge(df_hoods, on='Postal Code', how='left')
df_neighbourhoods.head()
```
